```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black image
image = np.zeros((500, 500, 3), dtype=np.uint8)

# Define the center and radius of the circle
center = (250, 250)
radius = 10

# Define the color of the circle (in BGR format)
color = (0, 255, 255)  # Yellow color

# Draw the circle on the image
cv2.circle(image, center, radius, color, -1)

# Display the image
cv2.imshow("Yellow Circle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

then fill the image with up to N yellow circles. the circles should not touch

```python
maxsize = 1024
minsize = 200

N=100

minradius = 5
maxradius = 20
```

```python
import cv2
import numpy as np

# Generate a random size for the image between maxsize and minsize
image_size = np.random.randint(minsize, maxsize+1)

radius = np.random.randint(minradius, maxradius+1)

# Create a black image with the generated size
image = np.zeros((image_size, image_size, 3), dtype=np.uint8)

# Generate random positions and radii for the circles
positions = np.random.randint(radius, image_size-radius, size=(N, 2))


# Define the color of the circles (in BGR format)
color = (0, 255, 255)  # Yellow color

# Draw the circles on the image
for i in range(N):
    center = tuple(positions[i])
    cv2.circle(image, center, radius, color, 5)


print("positions: ", positions)

# Display the image
cv2.imshow("Image with Yellow Circles", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def add_circle(image, center, circle):
    x, y = center
    x -= circle.shape[0] // 2
    y -= circle.shape[1] // 2
    image[x:(x+circle.shape[0]), y:(y+circle.shape[1])] += circle
    return image
```
